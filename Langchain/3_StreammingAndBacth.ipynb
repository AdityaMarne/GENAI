# Streaming

By displaying output progressively, even before a complete response is ready, streaming significantly improves user experience (UX), particularly when dealing with the latency of LLMs.

**streaming** is a key topic covered across different frameworks to improve user experience by displaying LLM output incrementally rather than waiting for the entire generation to complete.

### ** Streaming in LangChain **
* **Why it matters:** Unlike `model.invoke` which waits for the full response, `model.stream` yields output chunks as they are generated. This significantly improves the responsiveness of applications, especially for long-form content.
* **Implementation:** using `model.stream` to print chunks progressively.

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY')
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("groq:llama-3.1-8b-instant")
model
```

```python
model.invoke('Write a 200 words paragraph on Artificial Intelligence')
```

```python
for chunk in model.stream('Write a 200 words paragraph on Artificial Intelligence'):
    print(chunk.text, end="|", flush=True)
```

# Batch

In LangChain, **batch** is a powerful functionality used to process a collection of independent requests to an LLM in parallel. This approach significantly improves performance and reduces cost compared to processing requests sequentially.

**Key Features of Batch Processing:**
* **Efficiency:** By sending multiple queries (e.g., a list of questions) at once, the system processes them simultaneously rather than waiting for one to finish before starting the next.
* **Performance Optimization:** You can control the parallel execution using the `max_concurrency` configuration parameter, which determines how many calls are processed at the same time.
* **Use Case:** This is ideal for scenarios where you have a large set of inputs that don't depend on each other, allowing you to maximize throughput and minimize latency.

```python
model.batch(
    ["Hello",
     "What you are doing ?",
     "Write a 200 words paragraph on Artificial Intelligence"],
    config={
        "max_concurrency" : 5
    }
)
```

In *LangChain*, **max_concurrency** is a configuration parameter used within the `batch` method to control how many operations or LLM calls are executed in parallel.

**Key details:**
* **Purpose:** It allows you to manage the throughput of your requests, ensuring you don't exceed rate limits set by your LLM provider while still benefiting from parallel processing.
* **Usage:** When you pass a large list of inputs to `model.batch()`, setting `max_concurrency` determines the number of concurrent tasks. For example, if you have 10 inputs and set `max_concurrency=5`, the system will process the requests in two batches of five rather than all at once.
