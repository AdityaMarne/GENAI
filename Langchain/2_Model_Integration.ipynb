# Model Integration with OpenAi, Gemini, Groq

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
os.environ['GOOGLE_API_KEY'] = os.getenv('GOOGLE_API_KEY')
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

**`init_chat_model`** is a library utility in *LangChain* used to initialize and configure different types of LLMs (Large Language Models) in a unified way.

Key features of `init_chat_model` include:

* **Provider Agnosticism:** It provides a consistent interface to initialize models from various providers like *OpenAI*, *Google Gemini*, and *Groq* without needing to manually import specific class wrappers for each one.
* **Ease of Use:** Instead of complex initialization code, you can pass the model name directly, and the library handles the backend setup.
* **Flexibility:** It allows developers to switch between models (e.g., swapping from *GPT-4.1* to *Gemini 2.5 Flash*) by simply updating the model identifier string within the function.

Essentially, it simplifies the initialization process by acting as a streamlined entry point for connecting to your chosen AI models.

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-4.1")
model
```

```python
# invoke the model
response  = model.invoke('Hello how are you ?')

response
```
