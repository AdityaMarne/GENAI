# LangGraph Overview 

## What is LangGraph?

**LangGraph** is a **low-level orchestration framework** for building **stateful, long-running AI agents**.

Think of it like this:

* **LangChain** → Helps build AI applications.
* **LangGraph** → Controls **how the AI agent thinks, remembers, loops, and executes tasks**.

Unlike simple chains, LangGraph lets an agent:

* Remember previous information
* Pause and resume execution
* Ask humans for help
* Loop until a task is completed
* Recover after failures

It is designed specifically for **complex AI agents** rather than simple prompt-response applications. ([Docs by LangChain][1])

---

# Why LangGraph Exists

Traditional LangChain chains are mostly linear.

Example:

```
Input
   ↓
Prompt
   ↓
LLM
   ↓
Output
```

Real AI agents are more like:

```
User
   ↓
Reason
   ↓
Use Tool
   ↓
Analyze Result
   ↓
Need More?
   ↓
Yes → Repeat
No  → Final Answer
```

LangGraph is built to support these kinds of loops and decision-making. ([Docs by LangChain][1])

# When Should You Use LangGraph?

Use LangGraph if your application needs:

* Multi-step reasoning
* Multiple AI agents
* Tool calling
* Memory
* Long-running workflows
* Human approval
* Recovery after crashes

Examples:

* Customer support agent
* Coding assistant
* Research assistant
* Travel planner
* Autonomous AI systems
* RAG agents

# LangGraph vs LangChain

| LangChain              | LangGraph                         |
| ---------------------- | --------------------------------- |
| High-level framework   | Low-level orchestration framework |
| Easier for beginners   | More flexible                     |
| Good for simple agents | Good for production agents        |

The documentation recommends starting with LangChain agents if you're new, then moving to LangGraph when you need more control. ([Docs by LangChain][1])

# Core Benefits

## 1. Durable Execution ⭐

An agent can continue running even after failures.

Example:

```
Step 1 ✔
Step 2 ✔
Crash ❌

Restart

Continues from Step 3
```

Instead of restarting from the beginning, the agent resumes from where it stopped. ([Docs by LangChain][1])

---

## 2. Human-in-the-Loop

A person can interrupt the workflow, inspect or modify the agent's state, then allow it to continue.

Example:

```
Agent drafts contract

↓

Human reviews

↓

Approve

↓

Agent sends contract
```

Useful when human approval is required.

---

## 3. Comprehensive Memory

LangGraph supports both:

### Short-Term Memory

Information used during the current conversation.

Example:

```
User:
My name is Rahul.

Later:

AI:
Hello Rahul.
```

---

### Long-Term Memory

Information retained across different sessions.

Example:

```
Yesterday:
Favorite language = Python

Today:
AI remembers that preference.
```

---

## 4. Debugging

Works well with LangSmith to visualize:

* Graph execution
* State changes
* Node execution
* Errors
* Performance metrics

This makes debugging complex agents much easier. ([Docs by LangChain][1])

---

## 5. Production Ready

Designed for:

* Long-running agents
* Scalability
* Reliability
* Persistence
* Monitoring

# Hello World Example

```python
from langgraph.graph import (
    StateGraph,
    MessagesState,
    START,
    END
)

def mock_llm(state):
    return {
        "messages": [
            {
                "role": "ai",
                "content": "hello world"
            }
        ]
    }

graph = StateGraph(MessagesState)

graph.add_node(mock_llm)

graph.add_edge(START, "mock_llm")

graph.add_edge("mock_llm", END)

graph = graph.compile()

graph.invoke({
    "messages": [
        {
            "role": "user",
            "content": "hi!"
        }
    ]
})
```

### What this code does

* Creates a graph.
* Defines one node (`mock_llm`).
* Connects `START` → `mock_llm`.
* Connects `mock_llm` → `END`.
* Compiles the graph.
* Executes it with user input.
